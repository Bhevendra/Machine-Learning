```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt     # Import the library
```

```python
df = pd.read_csv('Position_Salaries.csv') #Import the data
```

```python
x = df.iloc[:,1:-1].values 
```

```python
y = df.iloc[:,-1].values
```

```python
from sklearn.linear_model import LinearRegression                #Training in Linear Regression Model
lin_reg = LinearRegression()
lin_reg.fit(x,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures            #Training in Polynomial Regression
poly_reg = PolynomialFeatures(degree=4)
x_poly = poly_reg.fit_transform(x)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(x_poly,y)
```

```python
plt.scatter(x,y,color='red')                                 #Visualizing the Linear Regression
plt.plot(x,lin_reg.predict(x),color='blue')
plt.title('Linear Regression')
plt.xlabel('Position')
plt.ylabel('salary')
plt.show()
```

```python
plt.scatter(x,y, color = 'red')                                     # Visualizing the Polynomial Regression)
plt.plot(x,lin_reg_2.predict(poly_reg.fit_transform(x)), color = 'blue')
plt.title('Polynomial Regression')
plt.xlabel('Position')
plt.ylabel('salary')
plt.show()
```

```python
x_grid = np.arange(min(x), max(x),0.1)
x_grid = x_grid.reshape(len(x_grid),1)
plt.scatter(x,y,color = 'red')
plt.plot(x_grid,lin_reg_2.predict(poly_reg.fit_transform(x_grid)), color='blue')
plt.title('Polynomial Regression1')
plt.xlabel('Position')
plt.ylabel('salary')
plt.show()
```
